```python
#데이터 전처리
import re
def remove_between_square_brackets(text):
    pattern='[^a-zA-z0-9가-힣,]'
    text=re.sub(pattern,'',text)
    #print(text)
    text = re.sub('[-=#/?:$\[\]}]', '', text)
    #print(text)
    str_list = text.split(',')
    return str_list

alist=[]
for i in range(len(lines)):
    alist.append(remove_between_square_brackets(lines[i]))
alist
```

```python
for i in range(len(alist)):
    if '' in alist[i]:
        print("[", i, "]", alist[i])
        alist[i].remove('')
```

```python
class Callback(CallbackAny2Vec):
    def __init__(self):
        self.epoch = 1
        self.training_loss = []

    def on_epoch_end(self, model):
        loss = model.get_latest_training_loss()
        if self.epoch == 1:
            current_loss = loss
        else:
            current_loss = loss - self.loss_previous_step
        print(f"Loss after epoch {self.epoch}: {current_loss}")
        self.training_loss.append(current_loss)
        self.epoch += 1
        self.loss_previous_step = loss
```

```python
f = open('./tag_remove_song2vec_alldata_model_train_epoch_250.dat', 'rb')
model=pickle.load(f)
#list = f.readlines()
#f.close()
print(model)
```

```python
#playlist 안의 노래들의 벡터의 mean값 구하기
def meanVectors(playlist):
    vec = []
    for song_id in playlist:
        try:
            vec.append(model.wv[song_id])
        except KeyError:
            continue
    return np.mean(vec, axis=0)
```

```python
#cols = ['tags', 'weight']
err = []
#playlist_tags = pd.DataFrame(columns=cols)
playlist_tags=[]
playlist_tags_weight=[]
for i in range(0,23015):
    playlist_vec=meanVectors(alist[i])
    if np.any(np.isnan(playlist_vec)):
        err.append(i)
        print("=================[ERR] in TRY", i)
        continue
    else:
        similar_tag=model.wv.similar_by_vector(playlist_vec, topn=10)
    #print(similar_songs)
    weight=[]
    song=[]
    for j in range(10):   
        song.append(similar_tag[j][0])
        weight.append(similar_tag[j][1])
    playlist_tags.append(song)
    playlist_tags_weight.append(weight)
    print("[DONE] in TRY", i)
```
